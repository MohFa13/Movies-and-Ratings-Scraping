```python
# imports
from bokeh.plotting import figure, show, output_notebook
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

# selenium
driver = webdriver.Chrome("C:\SeleniumDr\chromedriver.exe")
driver.get('https://www.imdb.com/registration/signin?ref=nv_generic_lgin&u=%2F')
driver.implicitly_wait(30)
form = driver.find_element(By.XPATH, '//*[@id="signin-options"]/div/div[1]/a[1]')
form.click()
username = driver.find_element(By.XPATH, '//*[@id="ap_email"]')
username.send_keys('xejeye3189@syswift.com')
password = driver.find_element(By.XPATH, '//*[@id="ap_password"]')
password.send_keys('TestProject123@')
login = driver.find_element(By.XPATH, '//*[@id="signInSubmit"]')
login.click()
driver.quit()


# crawler and regex
pages_crawled = []

page = requests.get('https://www.imdb.com/')
soup = BeautifulSoup(page.text, "html.parser")
links = soup.find_all('a')
for link in links:
    pages_crawled.append(link.get('href'))
url = re.findall(".[a-z]+.moviemeter", str(pages_crawled))
for i in range(0,len(url)):
    movie = "https://www.imdb.com/" + url[i]

    
# bs
titles = []
ratings = []

menu = requests.get(movie)
soup = BeautifulSoup(menu.text ,"html.parser")
title = soup.find_all('td', class_="titleColumn")

for td in title:
    titles.append(td.a.text)

rating = soup.find_all('strong')

for strong in rating:
    ratings.append(strong.text)

user_number = int(input("enter the size of the list (max is 50): "))


# Data visualization
d = {
    'title': titles[0:user_number],
    'rating': ratings[0:user_number]
}
df = pd.DataFrame(d)
output_notebook()
x = df['title'].index
y = df['rating'].values
p = figure(
    title="Movies Ratings",
    x_axis_label= "Movies",
    y_axis_label= "Rating"
)

p.line(x, y, legend='Movies Ratings of user choice')
show(p)
```
